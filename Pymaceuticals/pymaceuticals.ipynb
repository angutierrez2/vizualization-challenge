## Prepare the Data

```python
# Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import scipy.stats as sem
from scipy import stats
from pathlib import Path

# Study data files
mouse_metadata_path = Path("data/Mouse_metadata.csv")
study_results_path = Path("data/Study_results.csv")

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_study_data = study_results.merge(mouse_metadata, on="Mouse ID", how = "left")

# Display the data table for preview
combined_study_data.head()
```

```python
# Checking the number of mice
combined_study_data.groupby("Mouse ID").size().count()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint
duplicate = combined_study_data[combined_study_data.duplicated(["Mouse ID", "Timepoint"])]
duplicate["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID
combined_study_data[combined_study_data["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID
clean_combined_study_data = combined_study_data.drop(labels=[107, 137, 329, 360, 620, 681, 815, 869, 950, 1111, 1195, 1380, 1592], axis=0, inplace=False)
clean_combined_study_data.head()
```

```python
# Checking the number of mice in the clean DataFrame
clean_combined_study_data.groupby("Mouse ID").size().count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume
import numpy as np

capomulin = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Capomulin")]["Tumor Volume (mm3)"]
ceftamin = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Ceftamin")]["Tumor Volume (mm3)"]
infubinol = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Infubinol")]["Tumor Volume (mm3)"]
ketapril = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Ketapril")]["Tumor Volume (mm3)"]
naftisol = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Naftisol")]["Tumor Volume (mm3)"]
placebo = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Placebo")]["Tumor Volume (mm3)"]
propriva = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Propriva")]["Tumor Volume (mm3)"]
ramicane = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Ramicane")]["Tumor Volume (mm3)"]
stelasyn = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Stelasyn")]["Tumor Volume (mm3)"]
zoniferol = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Zoniferol")]["Tumor Volume (mm3)"]

mean_capomulin = np.mean(capomulin)
mean_ceftamin = np.mean(ceftamin)
mean_infubinol = np.mean(infubinol)
mean_ketapril = np.mean(ketapril)
mean_naftisol = np.mean(naftisol)
mean_placebo = np.mean(placebo)
mean_propriva = np.mean(propriva)
mean_ramicane = np.mean(ramicane)
mean_stelasyn = np.mean(stelasyn)
mean_zoniferol = np.mean(zoniferol)

median_capomulin = np.median(capomulin)
median_ceftamin = np.median(ceftamin)
median_infubinol = np.median(infubinol)
median_ketapril = np.median(ketapril)
median_naftisol = np.median(naftisol)
median_placebo = np.median(placebo)
median_propriva = np.median(propriva)
median_ramicane = np.median(ramicane)
median_stelasyn = np.median(stelasyn)
median_zoniferol = np.median(zoniferol)

var_capomulin = np.var(capomulin, ddof = 1)
var_ceftamin = np.var(ceftamin, ddof = 1)
var_infubinol = np.var(infubinol, ddof = 1)
var_ketapril = np.var(ketapril, ddof = 1)
var_naftisol = np.var(naftisol, ddof = 1)
var_placebo = np.var(placebo, ddof = 1)
var_propriva = np.var(propriva, ddof = 1)
var_ramicane = np.var(ramicane, ddof = 1)
var_stelasyn = np.var(stelasyn, ddof = 1)
var_zoniferol = np.var(zoniferol, ddof = 1)

sd_capomulin = np.std(capomulin, ddof = 1)
sd_ceftamin = np.std(ceftamin, ddof = 1)
sd_infubinol = np.std(infubinol, ddof = 1)
sd_ketapril = np.std(ketapril, ddof = 1)
sd_naftisol = np.std(naftisol, ddof = 1)
sd_placebo = np.std(placebo, ddof = 1)
sd_propriva = np.std(propriva, ddof = 1)
sd_ramicane = np.std(ramicane, ddof = 1)
sd_stelasyn = np.std(stelasyn, ddof = 1)
sd_zoniferol = np.std(zoniferol, ddof = 1)

sem_capomulin = capomulin.sem().round(6)
sem_ceftamin = ceftamin.sem().round(6)
sem_infubinol = infubinol.sem().round(6)
sem_ketapril = ketapril.sem().round(6)
sem_naftisol = naftisol.sem().round(6)
sem_placebo = placebo.sem().round(6)
sem_propriva = propriva.sem().round(6)
sem_ramicane = ramicane.sem().round(6)
sem_stelasyn = stelasyn.sem().round(6)
sem_zoniferol = zoniferol.sem().round(6)

# Assemble the resulting series into a single summary DataFrame
summary_statistics = pd.DataFrame({
    "Mean Tumor Volume": [mean_capomulin, mean_ceftamin, mean_infubinol, mean_ketapril, mean_naftisol, mean_placebo, mean_propriva, mean_ramicane, mean_stelasyn, mean_zoniferol],
    "Median Tumor Volume": [median_capomulin, median_ceftamin, median_infubinol, median_ketapril, median_naftisol, median_placebo, median_propriva, median_ramicane, median_stelasyn, median_zoniferol],
    "Tumor Volume Variance": [var_capomulin, var_ceftamin, var_infubinol, var_ketapril, var_naftisol, var_placebo, var_propriva, var_ramicane, var_stelasyn, var_zoniferol],
    "Tumor Volume Std. Dev.": [sd_capomulin, sd_ceftamin, sd_infubinol, sd_ketapril, sd_naftisol, sd_placebo, sd_propriva, sd_ramicane, sd_stelasyn, sd_zoniferol],
    "Tumor Volume Std. Err.": [sem_capomulin, sem_ceftamin, sem_infubinol, sem_ketapril, sem_naftisol, sem_placebo, sem_propriva, sem_ramicane, sem_stelasyn, sem_zoniferol]
    }, index = ["Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol", "Placebo", "Propriva", "Ramicane", "Stelasyn", "Zoniferol"])

# Formatting
summary_statistics.index.name = "Drug Regimen"

# Display Summary DataFrame
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
capomulin_last = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Capomulin")]["Timepoint"].iloc[-1]
ramicane_last = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Ramicane")]["Timepoint"].iloc[-1]
infubinol_last = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Infubinol")]["Timepoint"].iloc[-1]
ceftamin_last = clean_combined_study_data[(clean_combined_study_data["Drug Regimen"] == "Ceftamin")]["Timepoint"].iloc[-1]
capomulin_tumor = 31.02392294
ramicane_tumor = 33.56240217
infubinol_tumor = 60.91876652
ceftamin_tumor = 61.43389223

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_timepoint = pd.DataFrame({ "Last Timepoint": [capomulin_last, ramicane_last, infubinol_last, ceftamin_last],
                              "Tumor Volume": [capomulin_tumor, ramicane_tumor, infubinol_tumor, ceftamin_tumor]},
                              index = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
last_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
tumor_df = clean_combined_study_data.copy()
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_vol = [tumor_df[(tumor_df["Drug Regimen"] == "Capomulin")]["Tumor Volume (mm3)"],
             tumor_df[tumor_df["Drug Regimen"] == "Ramicane"]["Tumor Volume (mm3)"],
             tumor_df[(tumor_df["Drug Regimen"] == "Infubinol")]["Tumor Volume (mm3)"],
             tumor_df[(tumor_df["Drug Regimen"] == "Ceftamin")]["Tumor Volume (mm3)"]]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles_capomulin = tumor_vol[0].quantile([.25,.5,.75])
lowerq_capomulin = quartiles_capomulin[0.25]
upperq_capomulin = quartiles_capomulin[0.75]
iqr_capomulin = upperq_capomulin-lowerq_capomulin

quartiles_ramicane = tumor_vol[1].quantile([.25,.5,.75])
lowerq_ramicane = quartiles_ramicane[0.25]
upperq_ramicane = quartiles_ramicane[0.75]
iqr_ramicane = upperq_ramicane-lowerq_ramicane

quartiles_infubinol = tumor_vol[2].quantile([.25,.5,.75])
lowerq_infubinol = quartiles_infubinol[0.25]
upperq_infubinol = quartiles_infubinol[0.75]
iqr_infubinol = upperq_infubinol-lowerq_infubinol

quartiles_ceftamin = tumor_vol[3].quantile([.25,.5,.75])
lowerq_ceftamin = quartiles_ceftamin[0.25]
upperq_ceftamin = quartiles_ceftamin[0.75]
iqr_ceftamin = upperq_ceftamin-lowerq_ceftamin
    
# Locate the rows which contain mice on each drug and get the tumor volumes
rows_capomulin = tumor_df[(tumor_df["Drug Regimen"] == "Capomulin")]["Tumor Volume (mm3)"].count()
rows_ramicane = tumor_df[(tumor_df["Drug Regimen"] == "Ramicane")]["Tumor Volume (mm3)"].count()
rows_infubinol = tumor_df[(tumor_df["Drug Regimen"] == "Infubinol")]["Tumor Volume (mm3)"].count()
rows_ceftamin = tumor_df[(tumor_df["Drug Regimen"] == "Ceftamin")]["Tumor Volume (mm3)"].count()
tumor_vol_data = [rows_capomulin, rows_ramicane, rows_infubinol, rows_ceftamin]
    
# Determine outliers using upper and lower bounds
lower_bound_capomulin = round(lowerq_capomulin - (1.5*iqr_capomulin), 6)
upper_bound_capomulin = round(upperq_capomulin + (1.5*iqr_capomulin), 6)
print(f"Capomulin's potential outliers: {lower_bound_capomulin}, {upper_bound_capomulin}")

lower_bound_ramicane = round(lowerq_ramicane - (1.5*iqr_ramicane), 6)
upper_bound_ramicane = round(upperq_ramicane + (1.5*iqr_ramicane), 6)
print(f"Ramicane's potential outliers: {lower_bound_ramicane}, {upper_bound_ramicane}")

lower_bound_infubinol = round(lowerq_infubinol - (1.5*iqr_infubinol), 6)
upper_bound_infubinol = round(upperq_infubinol + (1.5*iqr_infubinol), 6)
print(f"Infubinol's potential outliers: {lower_bound_infubinol}, {upper_bound_infubinol}")

lower_bound_ceftamin = round(lowerq_ceftamin - (1.5*iqr_ceftamin), 6)
upper_bound_ceftamin = round(upperq_ceftamin + (1.5*iqr_ceftamin), 6)
print(f"Ceftamin's potential outliers: {lower_bound_ceftamin}, {upper_bound_ceftamin}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
capomulin_final = tumor_df[(tumor_df["Drug Regimen"] == "Capomulin")]["Tumor Volume (mm3)"]
ramicane_final = tumor_df[(tumor_df["Drug Regimen"] == "Ramicane")]["Tumor Volume (mm3)"]
infubinol_final = tumor_df[(tumor_df["Drug Regimen"] == "Infubinol")]["Tumor Volume (mm3)"]
ceftamin_final = tumor_df[(tumor_df["Drug Regimen"] == "Ceftamin")]["Tumor Volume (mm3)"]

final_df = pd.DataFrame({
    "Final Tumors": [capomulin_final, ramicane_final, infubinol_final, ceftamin_final]})

fig3, ax3 = plt.subplots()
ax3.set_ylabel('Final Tumor Volume (mm3)')
ax3.boxplot(final_df["Final Tumors"])
plt.xticks([1, 2, 3, 4], [treatments[0], treatments[1], treatments[2], treatments[3]])
plt.show()
```

## Correlation and Regression

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
mouse_weight = tumor_df[(tumor_df["Drug Regimen"] == "Capomulin")]["Weight (g)"]
average_tumor = tumor_df[(tumor_df["Drug Regimen"] == "Capomulin")]["Tumor Volume (mm3)"]

x_axis = mouse_weight
data = average_tumor
plt.scatter(x_axis, data, marker="o", facecolors="blue", edgecolors="white",
            s=50)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(x_axis, data)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
pe_slope, pe_int, pe_r, pe_p, pe_std_err = stats.linregress(x_axis, data)
pe_fit = pe_slope * x_axis + pe_int

plt.scatter(x_axis, data, marker="o", facecolors="blue", edgecolors="white",
            s=50)
plt.plot(x_axis,pe_fit, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
